**Categories Creation**

We created two categories in roder to continue with our analysis. 
* **First Category:** Level of physical activity
* **Second Category:** Device usage physical activity

These catergories would follow the following arguments:

**First Category**

* Sedentary: less than 6000 on average
* Active: between 6000 and 12000 on average
* Very active: more than 12000 on average

**Second Category**

* Low use: less than8 hours per day
* Normal use: between 8 adn 16 hours per day
* High use: more than 16 hours and up to 24 hours





```python
# Analysis 
# Categoreis
# Sedentary: less than 6000 on average
# Active: between 6000 and 12000 on average
# very active: more than 12000 on average

id_grp = df.groupby(['id'])
id_avg_step = id_grp['total_steps'].mean().sort_values(ascending=False)
id_avg_step = id_avg_step.to_frame()

conditions = [
    (id_avg_step <= 6000),
    (id_avg_step > 6000) & (id_avg_step < 12000),
    (id_avg_step >= 12000)
]
values = ['sedentery','active','very active']

id_avg_step['activity_level'] = np.select(conditions,values)

id_activity_level = id_avg_step['activity_level']

df['activity_level'] = [id_activity_level[c] for c in df['id']]

```

**Correlation between calories burned and steps**

What is the correlation between the amount of steps done, and the amount of calories burnt?

```python
# viz
# correlation between calories burned and steps

ax = sns.scatterplot(x = 'total_steps', y = 'calories', data=df)
plt.title('Correlations calories vs. Steps')

plt.tight_layout()
plt.show
```

In this scatterplot, we divided the dots by colors based on the activity_level category to visualize the groups represented in the dataset.

**Average number of steps per day**

What is the average number of steps per day?

```python
# average steps per day

day_of_week = ['Monday', 'Tuesday', 'Wednesday','Thursday','Friday','Saturday','Sunday']
fig, ax = plt.subplots(1,1, figsize=(8,5))

day_grp = df.groupby(['day_of_week'])
avg_daily_steps = day_grp['total_steps'].mean()
avg_steps = df['total_steps'].mean()

plt.title('Average steps per day')

plt.bar(avg_daily_steps.index, avg_daily_steps)

ax.set_xticks(range(len(day_of_week)))
ax.set_xticklabels(day_of_week)

ax.axhline(y=avg_daily_steps.mean(), color = 'blue', label='Avg daily steps')

ax.set_ylabel('Steps')
ax.set_xlabel('Day of Week')
plt.legend()

plt.show()
```

We can observe from the results that Tuesday, Wednesday, and Saturday are the days when users were most physically active, exceeding the average number of steps. Mondays, Fridays, and Sundays are slightly below the average but not significantly so. In contrast, Thursdays appear to be the least active day among the weekdays.

Based on this information, we can interpret that users are more active at the beginning of the week, with a decline midweek followed by a resurgence on Saturday.

**Percentage of activity in minutes**

What percentage of the time are people active?

```python
# Percentage of activity in minutes

very_active_mins = df['very_active_minutes'].sum()
fairly_active_mins = df['fairly_active_minutes'].sum()
lightly_active_mins = df['lightly_active_minutes'].sum()
sedentary_mins = df['sedentary_minutes'].sum()

slices = [very_active_mins,fairly_active_mins,lightly_active_mins,sedentary_mins]
labels = ['very_active_mins','fairly_active_mins','lightly_active_mins','sedentary_mins']

explode = [0,0,0,0.01]

plt.pie(slices, labels=labels, explode = explode, autopct = '%1.1f%%')

plt.title('Percentage of Activity Level in Minutes')

plt.show()
```

The pie chart reveals that users spend the majority of their time in a sedentary state, followed by one-sixth of the time engaged in light activity, and only 2% of the time actively exercising. However, this doesn't necessarily indicate laziness; rather, it's possible that the device is also tracking periods of sleep, contributing to the accumulation of sedentary minutes.

After analyzing the data, we found some insights that would help influence Bellabeat's marketing strategy.

# Notification on steps and lack of activity

Users have the option to set daily step goals using the device. By establishing a specific target for the number of steps they aim to achieve each day, users can better track and monitor their activity levels. The device then provides notifications to users, alerting them once they have successfully reached their set goal or if they are yet to meet it. This feature not will only encourages users to stay motivated and active throughout the day but also provides them with valuable feedback on their progress towards achieving their fitness objectives.

# Reward System

Not everyone has the motivation to do physical activity every day, but that doesn't mean it couldn't be fun. By implementing a reward system, people can track their progress by unlocking different levels once they have completed one they jump to the next one. Additionally, they can compete with other people in their area as well as with friends. 

By implementing a reward system and enabling competition among users, individuals are not only motivated to track their progress but also feel a sense of camaraderie with others. This shared experience of striving towards fitness goals fosters a supportive community environment, where users can exchange tips, encouragement, and even engage in friendly competition. As users unlock different levels and achieve milestones, they not only celebrate their own accomplishments but also cheer on others within the community, further strengthening bonds and fostering a sense of belonging.

Ultimately, what may have started as a solitary pursuit of personal fitness goals evolves into a vibrant and interconnected community of individuals striving towards better health and well-being together.
